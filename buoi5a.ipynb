```python
import pandas as pd
```

```python
df = pd.read_csv(filepath_or_buffer="E:\Python\cities_population.csv", header=0, encoding = 'utf8')
df
```

## 1,  Sắp xếp và in ra 10 thành phố có dân số lớn nhất & 10 thành phố có dân số nhỏ nhất

```python
df_1 = df[["City","Population"]]
df_1['Population'] = df_1['Population'].apply(lambda x: x.replace(',',''))
df_1["Population"]= df_1["Population"].astype(int)
print(type(df_1["Population"]))

sort_df_1 = df_1.sort_values("Population", ascending= False)
top_10 = sort_df_1[:10]
inverse_top10 = sort_df_1[-10:]
print(top_10)
print(inverse_top10)
```

## 3, In ra Top 5 quốc gia có nhiều thành phố xuất hiện trong bảng này nhất

```python
df_3 = df[["City","Country"]]
df_3['Country']= df_3['Country'].str.replace('*','')
df_3['Country']= df_3['Country'].str.replace(' ','')

sort_country = df_3['Country'].value_counts()
top_5_country = sort_country[:5]

print(top_5_country)
```

## 4,  In ra các thành phố có dân số & diện tích đều nằm trong Top 20


```python
df_4 = df[['City','Population','Area KM2']]
df_4['Area KM2'] = df_4['Area KM2'].str.replace('7-May','7')
df_4['Population']= df_4['Population'].str.replace(',','')
df_4['Population']= df_4['Population'].astype(int)
df_4['Area KM2']= df_4['Area KM2'].astype(int)


sort_population = df_4.sort_values(['Population'], ascending=False)
sort_area = df_4.sort_values(['Area KM2'], ascending=False)
top20_population = sort_population[:20]
top20_area = sort_area[:20]
print(sort_population)
print(top20_population,'\n')
print(top20_area)
a= top20_population['Population'].min()
ontop20 = top20_area.loc[top20_area.Population > a]
ontop20

```

## 5, : Thống kê mật độ dân số theo quốc gia


```python
df_5 = df[['Population','Area KM2', 'Country']]
df_5['Area KM2'] = df_5['Area KM2'].str.replace('7-May','7')
df_5['Population']= df_5['Population'].str.replace(',','')
df_5['Population']= df_5['Population'].astype(int)
df_5['Area KM2']= df_5['Area KM2'].astype(int)
df_5['Country']= df_5['Country'].str.replace('*','')
df_5['Country']= df_5['Country'].str.replace(' ','')


#group_country = df_5.groupby("Country").apply(print)
group_country = df_5.groupby("Country")['Population','Area KM2'].sum()
group_country["Density"] = group_country["Population"]/group_country["Area KM2"]
print(group_country)


```
